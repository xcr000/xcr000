```python
import pandas as pd
import numpy as np
from scipy.stats import skew
from scipy.stats import kurtosis

JPM_df = pd.read_csv('JPM.csv')

#Use adjusted closing price:
prices = JPM_df['Adj Close']

#Calculate one day returns and its statistics:
returns = prices/prices.shift(1) - 1

mean = np.mean(returns)
std = np.std(returns)
skew = returns.skew()
kurtosis = returns.kurtosis()

p = [0.01, 0.03, 0.05, 0.10, 0.25, 0.5, 0.75, 0.90, 0.95, 0.97, 0.99]
R_percentile = returns.quantile(p)

print("The mean daily return is ", mean, '\n',
      "The standard deviation of daily return is ", std,'\n',
      "The skewness of daily return is " ,skew,'\n',
      "The kurtosis of daily return is " ,kurtosis,'\n',
      "The percentiles are " ,R_percentile)

#Calculate annualized data and sharpe ratio
annualizedMeanR = mean*251
annualizedStdR = std*np.sqrt(251)
SharpeRatio = annualizedMeanR/annualizedStdR

#Calculate YTD return (from Jan 3rd 2023 to Oct 20 2023)
start_2023 = JPM_df[(JPM_df.Date == "2023-01-03")]['Adj Close']
todate_2023 = JPM_df[(JPM_df.Date == "2023-10-20")]['Adj Close']
YTD_return = float(todate_2023) / float(start_2023) -1

print("The annualized return is ",annualizedMeanR, '\n',
      "The Year-to-date (YTD) return is ", YTD_return, '\n',
      "The annualized Sharpe Ratio is ",  SharpeRatio)

```
